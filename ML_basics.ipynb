```python
def two_class_knn_plot(x,y,n_neighbors,weights,x_test,y_test):
    X_mat=x
    Y_mat=y
    
    cmap_light = ListedColormap(['#FFFFAA', '#AAFFAA', '#AAAAFF','#EFEFEF'])
    cmap_bold  = ListedColormap(['#FFFF00', '#00FF00', '#0000FF','#000000'])
    knn=KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_mat,Y_mat)
    x_min= X_mat[:,0].min()-1
    x_max= X_mat[:,0].max()+1
    y_min= X_mat[:,1].min()-1
    y_max= X_mat[:,1].max()+1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),np.arange(y_min, y_max, 0.01))
    pred=knn.predict(np.c_[xx.ravel(), yy.ravel()]) 
    pred=pred.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, pred, cmap=cmap_light)

    # Plot training points
    plt.scatter(X_mat[:, 0], X_mat[:, 1], s=50, c=y, cmap=cmap_bold, edgecolor = 'black')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())

    title = "Neighbors = {}".format(n_neighbors)
    if (x_test is not None):
        train_score = knn.score(X_mat, Y_mat)
        test_score  = knn.score(x_test, y_test)
        title = title + "\nTrain score = {:.2f}, Test score = {:.2f}".format(train_score, test_score)

    patch0 = mpatches.Patch(color='#FFFF00', label='class 0')
    patch1 = mpatches.Patch(color='#000000', label='class 1')
    plt.legend(handles=[patch0, patch1])
    plt.xlabel('Feature 0')
    plt.ylabel('Feature 1')
    plt.title(title)
    plt.show()
    
```

```python
from sklearn.datasets import  make_classification,make_regression
X1,Y1=make_classification(n_samples=100,n_features=2,n_redundant=0,n_informative=2,class_sep=0.5,random_state=0)
X2,Y2=make_regression(n_samples=100,n_features=1,n_informative=1,bias=150,noise=30,random_state=0)

x1_train,x1_test,y1_train,y1_test=train_test_split(X1,Y1,random_state=0)
two_class_knn_plot(x1_train,y1_train, 1,'uniform',x1_test,y1_test)
two_class_knn_plot(x1_train,y1_train, 3,'uniform',x1_test,y1_test)
two_class_knn_plot(x1_train,y1_train, 11,'uniform',x1_test,y1_test)
```

```python
#Ridge Regression (Used for less number of features)
```

##Basic formula for multiple categories regression is
= Sum((y-(wx+b))^2) + a*Sum(wi ^ 2)....where i=range(1,y)

```python
##Dataset for Ridge Regression

def generate_data(m=100, n=20, sigma=5):
    #"Generates data matrix X and observations Y."
    np.random.seed(1)
    beta_star = np.random.randn(n)
    # Generate an ill-conditioned data matrix
    X = np.random.randn(m, n)
    # Corrupt the observations with additive Gaussian noise
    Y = X.dot(beta_star) + np.random.normal(0, sigma, size=m)
    return X, Y

m = 100
n = 20
sigma = 5

X_R, Y_R = generate_data(m, n, sigma)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
#1/n(sum(y^-y))
x3_train,x3_test,y3_train,y3_test=train_test_split(X_R,Y_R,random_state=0)
alp=0.0
for alp in range(int(1.0),int(50.0)):
    lr_ridge=Ridge(alpha=alp)
    lr_ridge.fit(x3_train,y3_train)
    y_pred=lr_ridge.predict(x3_train)
    print(mean_squared_error(y_train,y_pred)) # Finds bias value
    print("for value of alpha equals {}".format(alp))
    print(lr_ridge.intercept_)
    print(lr_ridge.coef_)
    print(lr_ridge.score(x3_train,y3_train))
    print(lr_ridge.score(x3_test,y3_test))


```

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn import metrics

dataset=load_iris()
x,y=dataset.data,dataset.target
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
svc=SVC(kernel='linear',C=1.0)
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
svc.score(x_test,y_test)
print(metrics.confusion_matrix(y_test,y_pred))
print(metrics.classification_report(y_test,y_pred))
```
